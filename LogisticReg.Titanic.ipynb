```python
test["CabinAdv"]=test.Cabin.apply(lambda x: str(x)[0])

test["TicketNum"]=test.Ticket.apply(lambda x:1 if x.isnumeric() else 0)

test["NameTitle"]=test.Name.apply(lambda x:x.split(".")[0].split(",")[1].strip())

test.Age=test.Age.fillna(test.Age.mean())
test.Fare=test.Age.fillna(test.Fare.mean())
```

```python
combined=pd.concat([train,test],axis=0)
```

```python
combined
```

```python
combined.select_dtypes("object").columns
```

```python
newcomb= pd.get_dummies(combined, columns=['Sex', 'Embarked', 'CabinAdv', 'NameTitle'],drop_first=True)
```

```python
newtrain=newcomb.iloc[:889,:]
newtest=newcomb.iloc[889:,:]
```

```python
newtest.drop(["Survived"],axis=1,inplace=True)
```

```python
newtest
```

```python
X_train=newtrain.drop(["Survived"],axis=1)
Y_train=newtrain.Survived
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression(solver='lbfgs', max_iter=1000)
logmodel.fit(X_train,Y_train)
```

```python
predictions=logmodel.predict(newtest)
```

```python
predictions
```

```python
Ypredi=pd.DataFrame(predictions)
subd=pd.read_csv("titanic//gender_submission.csv")


# datasets.to_csv("gender_submission.csv",index=False)
```
